```python
import pandas as pd
```

```python
#from C:\Users\AvaStore\OneDrive - McGill University\NI-CG Folder\MITACS- Assomption Sud-Longue-Pointe\Lit Review\tables on mitacs folder\used
df= pd. read_excel ("Base for InteractiveMapTest.xlsx", header=[0, 1, 2])
for col in df.columns:
  print(col)
```

```python
df = df.set_index(('metadata', 'metadata', 'Cities'))
df_long = df.stack(level=[0, 1, 2]).reset_index()
df_long.columns = ['City', 'Strategy', 'SubStrategy', 'Intervention', 'Implemented']
print (df_long)
```

```python
df_long.to_excel('Cities&Strategies.xlsx', index=False)
```

```python
city_continent_df = pd.read_excel("Base for InteractiveMapTest.xlsx",sheet_name= 'continent', header=[0])
city_continent_df.columns = ['Continent', 'City', 'Country']
print (city_continent_df)
```

```python
intervention_counts = df_long.groupby('City').size().reset_index(name='Total_Interventions')
print (intervention_counts)
```

```python
df_long = df_long.merge(city_continent_df, on='City', how='left')
print (df_long)
```

```python
for col in df_long.columns:
  print(col)
```

```python
import plotly.express as px

fig = px.sunburst(
    df_long,
    path=['Continent', 'Country', 'City', 'Strategy', 'SubStrategy', 'Intervention'],
    values=None
)

fig.show()
```
